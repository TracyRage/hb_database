```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('PAH_dioxygenase_a.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,45) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=61, fill="steelblue", alpha=.25) +
geom_cladelabel(node=61, label='Sphingomonas, ahdA1a', geom='label',
               offset.text=15, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE)
ggsave('ahdA1a_tree.jpeg', scale=1.5)
ggsave('ahdA1a_tree.pdf', scale=1.5)
```
