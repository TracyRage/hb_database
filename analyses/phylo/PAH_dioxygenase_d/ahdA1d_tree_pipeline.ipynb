```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('PAH_dioxygenase_d.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,55) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=51, fill="steelblue", alpha=.25) +
geom_cladelabel(node=51, label='Sphingomonas, ahdA1d', geom='label',
               offset.text=20, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE)
ggsave('ahdA1d_tree.jpeg', scale=1.5)
ggsave('ahdA1d_tree.pdf', scale=1.5)
```
