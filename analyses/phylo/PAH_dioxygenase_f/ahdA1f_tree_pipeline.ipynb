```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('PAH_dioxygenase_f.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,50) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=73, fill="steelblue", alpha=.25) +
geom_cladelabel(node=73, label='Sphingobium, ahdA1f', geom='label',
               offset.text=17, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) 

ggsave('ahdA1f_tree.jpeg', scale=1.5)
ggsave('ahdA1f_tree.pdf', scale=1.5)
```
