```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Ethylbenzene_dehydrogenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,30) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=95, fill="steelblue", alpha=.25) +
geom_cladelabel(node=95, label='Azoarcus, ebdA1', geom='label',
               offset.text=15, fill='steelblue', alpha=.5, color='white',
                align=TRUE, barsize=0, hjust=0, fontsize=4)

ggsave('ebdA1_tree.jpeg', scale=1.5)
ggsave('ebdA1_tree.pdf', scale=1.5)
```
