```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Toluene_4-monooxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,65) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=70, fill="steelblue", alpha=.25) +
geom_cladelabel(node=70, label='Superclade, tmoA', geom='label',
               offset.text=17, fill='steelblue', alpha=1,
                 hjust=1.5, barsize=0, color='white', align=TRUE) 
ggsave('tmoA_tree.pdf', scale=1.5)
ggsave('tmoA_tree.jpeg', scale=1.5)
```
