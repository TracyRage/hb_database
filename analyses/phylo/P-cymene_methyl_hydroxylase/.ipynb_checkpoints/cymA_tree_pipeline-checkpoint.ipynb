```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('P-cymene_methyl_hydroxylase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,50) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=81, fill="steelblue", alpha=.25) +
geom_cladelabel(node=62, label='Pseudomonas, cymA', geom='label',
               offset.text=20, fill='steelblue', alpha=0.5,
                 hjust=2, barsize=0, color='white', align=TRUE)
ggsave('cymA_tree.jpeg', scale=1.5)
ggsave('cymA_tree.pdf', scale=1.5)
```
