```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Diphenyl_ether_dioxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,30) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=78, fill="steelblue", alpha=.5) +
geom_cladelabel(node=78, label='Sphingobium, dpeA1', geom='label',
               offset.text=15, fill='steelblue', alpha=.5, color='white',
                align=TRUE, fontsize=4, hjust=0)

ggsave('dpeA1_tree.pdf', scale=1.5)
ggsave('dpeA1_tree.jpeg', scale=1.5)
```
