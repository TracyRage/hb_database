```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Dibenzofuran_dioxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,35) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=85, fill="steelblue", alpha=.5) +
geom_cladelabel(node=85, label='Rhodococcus, dbfA1', geom='label',
               offset.text=22, fill='steelblue', alpha=.5,
                 hjust=0.5, barsize=0, color='white', align=TRUE, fontsize=4)

ggsave('dbfA1_tree.pdf', scale=1.5)
ggsave('dbfA1_tree.jpeg', scale=1.5)
```
