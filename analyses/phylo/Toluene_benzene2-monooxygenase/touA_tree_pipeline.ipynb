```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Toluene_benzene2-monooxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,50) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=21, fill="steelblue", alpha=.25) +
geom_cladelabel(node=21, label='Pseudomonas, touA', geom='label',
               offset.text=17, fill='steelblue', alpha=0.5,
                 hjust=-.5, barsize=0, color='white', align=TRUE) 
ggsave('touA_tree.jpeg', scale=1.5)
ggsave('touA_tree.pdf', scale=1.5)
```
