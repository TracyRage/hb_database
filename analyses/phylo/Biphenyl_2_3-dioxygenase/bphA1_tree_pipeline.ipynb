```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Biphenyl_2_3-dioxygenase.trimmed.contree')
```

```r
df  <- tbl_df(tree)
```

```r
# Cheack nodes qualified to be a true clade
filtered  <- tree_df  %>% 
filter(!is.na(UFboot) & UFboot >= 98 & SH_aLRT >= 80)
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,70) +
geom_tiplab(aes(angle=angle), size=2) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=107, fill="darkgreen", alpha=.25) +
geom_hilight(node=154, fill="firebrick", alpha=.5) +
geom_hilight(node=134, fill="steelblue", alpha=.5) +
geom_cladelabel(node=107, label='Superclade, bphA1', geom='label',
               offset.text=20, fill='darkgreen', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=154, label='Pseudomonas, bphA1', geom='label',
               offset.text=20, fill='firebrick', alpha=.5, color='white',
                align=TRUE, barsize=0, hjust=-0.5) +
geom_cladelabel(node=154, label='E. coli, bphA1', geom='label',
               offset.text=25, fill='steelblue', alpha=.5, color='white',
                hjust=5, barsize=0)

ggsave('bphA1_tree.pdf', scale=1.5)
ggsave('bphA1_tree.jpeg', scale=1.5)
```
