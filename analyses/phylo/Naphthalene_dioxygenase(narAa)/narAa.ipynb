```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Naphthalene_dioxygenase(narAa).trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,35) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=69, fill="darkgreen", alpha=.25) +
geom_hilight(node=77, fill="steelblue", alpha=.25) +
geom_cladelabel(node=69, label='Rhodococcus, narAa', geom='label',
               offset.text=20, fill='darkgreen', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=77, label='Mycobacterium, narAa', geom='label',
               offset.text=10, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) 
ggsave('narAa_tree.jpeg', scale=1.5)
ggsave('narAa_tree.pdf', scale=1.5)
```
