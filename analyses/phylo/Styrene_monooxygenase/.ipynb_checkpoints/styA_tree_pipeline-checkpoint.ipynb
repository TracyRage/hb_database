```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Styrene_monooxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,75) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=91, fill="steelblue", alpha=.25) +
geom_hilight(node=87, fill="firebrick", alpha=.25) +
geom_cladelabel(node=62, label='Superclade, styA', geom='label',
               offset.text=17, fill='steelblue', alpha=1,
                 hjust=1.5, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=62, label='Mycobacteroides, styA', geom='label',
               offset.text=20, fill='firebrick', alpha=1,
                 hjust=-2.5, barsize=0, color='white', align=TRUE) 
ggsave("styA_tree.jpeg", scale=1.5)
ggsave("styA_tree.pdf", scale=1.5)
```
