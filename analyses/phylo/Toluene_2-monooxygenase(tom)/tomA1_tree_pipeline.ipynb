```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Toluene_2-monooxygenase(tom).trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,25) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=23, fill="steelblue", alpha=.25) +
geom_hilight(node=15, fill="firebrick", alpha=.25) +
geom_hilight(node=1, fill="darkgreen", alpha=.25) +
geom_cladelabel(node=23, label='Clade 1, tomA1', geom='label',
               offset.text=17, fill='steelblue', alpha=0.5,
                 hjust=.3, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=15, label='Clade 2, tomA1', geom='label',
               offset.text=17, fill='firebrick', alpha=0.5,
                 hjust=2, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=15, label='Clade 3, tomA1', geom='label',
               offset.text=17, fill='darkgreen', alpha=0.5,
                 hjust=.3, barsize=0, color='white', align=TRUE)
ggsave('tomA1_tree.jpeg', scale=1.5)
ggsave('tomA1_tree.pdf', scale=1.5)
```
