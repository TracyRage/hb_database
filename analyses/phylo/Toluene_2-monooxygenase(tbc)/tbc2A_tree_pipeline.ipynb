```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Toluene_2-monooxygenase(tbc).trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,20) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=20, fill="steelblue", alpha=.25) +
geom_cladelabel(node=20, label='Burkholderia, tbc2A', geom='label',
               offset.text=5, fill='steelblue', alpha=0.5,
                 hjust=-.2, barsize=0, color='white', align=TRUE) 
ggsave('tbc2A.jpeg', scale=1.5)
ggsave('tbc2A.pdf', scale=1.5)
```
