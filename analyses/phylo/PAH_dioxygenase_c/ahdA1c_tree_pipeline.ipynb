```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('PAH_dioxygenase_c.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,40) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=56, fill="steelblue", alpha=.25) +
geom_cladelabel(node=56, label='Sphingomonas, ahdA1c', geom='label',
               offset.text=13, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) 
ggsave("ahdA1b_tree.jpeg", scale=1.5)
ggsave("ahdA1b_tree.pdf", scale=1.5)
```
