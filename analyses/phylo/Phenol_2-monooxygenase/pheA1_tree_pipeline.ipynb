```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Phenol_2-monooxygenase.trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,50) +
geom_tiplab(aes(angle=angle), size=3) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=81, fill="steelblue", alpha=.25) +
geom_hilight(node=78, fill="firebrick", alpha=.25) +
geom_cladelabel(node=81, label='Superclade, pheA1', geom='label',
               offset.text=17, fill='steelblue', alpha=0.5,
                 hjust=1.5, barsize=0, color='white', align=TRUE) +
geom_cladelabel(node=62, label='Haloferax, pheA1', geom='label',
               offset.text=20, fill='firebrick', alpha=0.5,
                 hjust=-.5, barsize=0, color='white', align=TRUE) 
ggsave('pheA1_tree.jpeg', scale=1.5)
ggsave('pheA1_tree.pdf', scale=1.5)
```
