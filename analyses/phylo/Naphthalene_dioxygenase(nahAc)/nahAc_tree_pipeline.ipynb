```r
library('dplyr')
library('treeio')
library('ggtree')
```

```r
tree  <- read.iqtree('Naphthalene_dioxygenase(nahAc).trimmed.contree')
```

```r
t  <- ggtree(tree, branch.length='none', layout='circular') + xlim(NA,35) +
geom_tiplab(aes(angle=angle), size=2.5) +
geom_point2(aes(label=label, subset=!is.na(UFboot) & UFboot >= 95 & SH_aLRT >= 80)) +
geom_hilight(node=77, fill="steelblue", alpha=.25) +
geom_cladelabel(node=77, label='Pseudomonas, nahAc', geom='label',
               offset.text=15, fill='steelblue', alpha=0.5,
                 hjust=-0.5, barsize=0, color='white', align=TRUE)
ggsave('nahAc_tree.pdf', scale=1.5)
ggsave('nahAc_tree.jpeg', scale=1.5)
```
